#Final data preprocessing

As a final pre processing step, let us employ Z-score normalization on the word count and char count fields.

```python
mean_word_count = df['word_count'].mean()
std_word_count = df['word_count'].std()
df['word_count'] = (df['word_count'] - mean_word_count) / std_word_count
mean_char_count = df['char_count'].mean()
std_char_count = df['char_count'].std()
df['char_count'] = (df['char_count'] - mean_char_count) / std_char_count
print(mean_word_count)
print(std_word_count)
print(mean_char_count)
print(std_char_count)
```

#Training a Linear Regression model

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

X = df.drop(columns = ['id', 'date', 'likes', 'log_likes', 'content', 'username', 'media', 'inferred company', 'has_media', 'datetime', 'hour', 'day_of_week', 'media_type'])
X.head(10)
```

```python
print(X.shape)
```

```python
y= df['log_likes']
y.head(10)
```

Let's split the data into train, dev and test to compare between different models

```python
X_temp, X_test, y_temp, y_test = train_test_split(X, y, test_size=0.1, random_state=42) # 10% of total data in test set since it is a large dataset
X_train, X_dev, y_train, y_dev = train_test_split(X_temp, y_temp, test_size=0.22, random_state=42) # approximately 20% of total data in dev set
```

Training without regularization first

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

Prediction and Evaluation

```python
y_pred = model.predict(X_dev)
from sklearn.metrics import mean_squared_error, r2_score
print("MSE:", mean_squared_error(y_dev, y_pred))
print("R²:", r2_score(y_dev, y_pred))
```

Let's try regularization to see if the model overfit

```python
from sklearn.linear_model import Ridge
model = Ridge(alpha=1.0)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_dev)
print("MSE:", mean_squared_error(y_dev, y_pred))
print("R²:", r2_score(y_dev, y_pred))
```

The model performs just slightly better. Let us try to visualize what the problem may be

```python
import matplotlib.pyplot as plt
residuals = y_dev - y_pred
plt.hist(residuals, bins=30)
plt.title("Residuals Distribution")
```

Let us now try out a different model, a DT based model, with bagging, Random Forests

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(
    n_estimators=200,
    max_depth=25,
    min_samples_split=3,
    random_state=42,
    n_jobs = -1
)

rf.fit(X_train, y_train)

y_dev_pred = rf.predict(X_dev)

print("MSE:", mean_squared_error(y_dev, y_dev_pred))
print("R²:", r2_score(y_dev, y_dev_pred))
```

Let's train yet another model, this time XGBoost

```python
import xgboost as xgb
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import r2_score, mean_squared_error
import numpy as np

xgb_model = xgb.XGBRegressor(
    objective='reg:squarederror',
    tree_method='gpu_hist',
    predictor='gpu_predictor',
    eval_metric='rmse',
    random_state=42
)

param_dist = {
    'n_estimators': [200, 400, 600],
    'max_depth': [4, 6, 8, 10],
    'learning_rate': [0.01, 0.05, 0.1],
    'subsample': [0.6, 0.8, 1.0],
    'colsample_bytree': [0.6, 0.8, 1.0],
    'min_child_weight': [1, 3, 5],
    'gamma': [0, 0.1, 0.3, 0.5],
    'reg_alpha': [0, 0.1, 0.5],
    'reg_lambda': [1, 3, 5]
}

random_search = RandomizedSearchCV(
    estimator=xgb_model,
    param_distributions=param_dist,
    n_iter=25,
    scoring='r2',
    cv=3,
    verbose=1,
    random_state=42,
    n_jobs=-1
)

random_search.fit(X_train, y_train)

best_model = random_search.best_estimator_

y_pred = best_model.predict(X_dev)
r2 = r2_score(y_dev, y_pred)
mse = mean_squared_error(y_dev, y_pred)

print(f"Best R² on dev set: {r2:.4f}")
print(f"MSE on dev set: {mse:.4f}")
print("Best Params:", random_search.best_params_)
```

```python
import joblib

joblib.dump(best_model, '/content/drive/MyDrive/xgboost_best_model.pkl')
```

Tying a neural network, the below code is generated by ChatGPT, and tweaked by me for optimal performance

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score, mean_squared_error

# Normalize features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_dev_scaled = scaler.transform(X_dev)

# Build the model
model = Sequential([
    Dense(512, activation='relu', input_shape=(X_train.shape[1],)),
    Dropout(0.3),
    Dense(256, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(1e-4)),
    Dropout(0.2),
    Dense(128, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(1e-4)),
    Dropout(0.2),
    Dense(32, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(1e-4)),
    Dropout(0.2),
    Dense(16, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(1e-4)),
    Dropout(0.2),
    Dense(1)  # Output layer for regression
])

# Compile the model
model.compile(
    optimizer='adam',
    loss='mse',
    metrics=[tf.keras.metrics.RootMeanSquaredError()]
)

# Early stopping to prevent overfitting
early_stop = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

# Train the model
history = model.fit(
    X_train_scaled, y_train,
    validation_data=(X_dev_scaled, y_dev),
    epochs=100,
    batch_size=1024,
    callbacks=[early_stop],
    verbose=1
)

# Predict and evaluate
y_pred = model.predict(X_dev_scaled)
r2 = r2_score(y_dev, y_pred)
mse = mean_squared_error(y_dev, y_pred)

print(f"R² on dev set: {r2:.4f}")
print(f"MSE on dev set: {mse:.4f}")
```

```python
y_pred_nn = y_pred = model.predict(X_dev_scaled)
xgb_model = joblib.load('/content/drive/MyDrive/xgboost_best_model.pkl')
y_pred_xgb = xgb_model.predict(X_dev)

y_pred = 0.3*y_pred_nn.flatten() + 0.7*y_pred_xgb.flatten()

r2 = r2_score(y_dev, y_pred)
mse = mean_squared_error(y_dev, y_pred)

print(f"R² on dev set: {r2:.4f}")
print(f"MSE on dev set: {mse:.4f}")
```
